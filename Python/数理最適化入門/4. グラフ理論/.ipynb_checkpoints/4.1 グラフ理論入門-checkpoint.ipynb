# グラフとは

```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()
# 頂点リスト
vlist = [1,2,3,4]
# 枝リスト
elist = [(1,2), (1,3), (2,3),(2,4),(3,4)]
G.add_nodes_from(vlist)
G.add_edges_from(elist)
nx.draw_networkx(G, node_color='lightgray', node_size=800)
plt.axis('off')
#plt.savefig('../fig/graph1.eps')
plt.show()
```

```python
print('Gの頂点のリスト:', G.nodes())
print('Gの頂点の数：', G.number_of_nodes())
print('Gの頂点:1に隣接する頂点のリスト:',
              [v for v in nx.all_neighbors(G,1)])
```

```python
# DiGraphとすると有向グラフになる
DG=nx.DiGraph()
vlist = [1,2,3,4]
elist = [(1,2), (1,3), (2,3),(2,4),(3,4)]
DG.add_nodes_from(vlist)
DG.add_edges_from(elist)
nx.draw_networkx(DG, node_color='lightgrey', node_size=800)
plt.axis('off')
#plt.savefig('../fig3/digraph1.eps')
plt.show()
```

```python
print('DGの枝のリスト:', DG.edges())
print('DGの枝の数：', DG.number_of_edges())
```

```python
# 隣接行列と接続行列
# 隣接行列：頂点同士がつながっていれば1、複数つながっていればその数になる
# 接続行列：頂点がエッジに接続していれば1、そうでなければ0になる
G = nx.MultiGraph()
G.add_edges_from([(1,2),(1,3),(3,1),(2,3),(2,2)])
A = nx.adjacency_matrix(G)
M = nx.incidence_matrix(G)

nx.draw_networkx(G, node_color='lightgray', node_size=800)
plt.axis('off')
#plt.savefig('../fig/graph1.eps')
plt.show()

print('A =', A.toarray())
print('M =', M.toarray())
```

```python
# 有向グラフの場合は、隣接行列は左右対称ではなくなる
# 接続行列は始点ならば-1、終点ならば+1
G = nx.MultiDiGraph()
G.add_edges_from([(1,2),(1,3),(3,1),(2,3),(2,3),(2,2)])
A = nx.adjacency_matrix(G)
M = nx.incidence_matrix(G, oriented=True)
print('A =', A.toarray())
print('M =', M.toarray())
```

# 完全グラフ

```python
G = nx.complete_graph(5) 
# 隣接する点同士は引き合い、隣接していない点同士は反発するという力学系を考え、平衡状態を反復で計算
p = nx.spring_layout(G, iterations=100)
nx.draw_networkx(G,pos=p,node_color='lightgrey',node_size=300)
plt.axis('off')
plt.show()
```

# 完全2部グラフ

```python
# このようなグラフを完全2部グラフという
m,n = 3,4
G = nx.complete_bipartite_graph(m,n)
nx.draw_networkx(G,pos=nx.random_layout(G),
                 node_color='lightgrey', node_size=500)
plt.axis('off')
plt.show()
```

```python
# 上記だと分かりにくいのでわかりやすく描画
# 2グループ(X,Y)に分けて、どの枝もXY両方の点に接続している状態
m,n = 3,4
G=nx.complete_bipartite_graph(m,n)
p = {}

# 座標を指定
for i in range(m):
    p[i] = (0,i)
for i in range(n):
    p[m+i] = (1,i)
nx.draw_networkx(G,pos=p,node_color='lightgrey',node_size=500)
plt.axis('off')
plt.show()
```

# サイクルグラフ

```python
n = 10
G = nx.cycle_graph(n)
p = nx.circular_layout(G)
nx.draw_networkx(G, pos=p,
                 node_color='lightgrey', node_size=500)
plt.axis('off')
plt.show()
```

# 星グラフ

```python
n = 10
G = nx.star_graph(n)
p = nx.spring_layout(G, iterations=100)
nx.draw_networkx(G, pos=p,
                 node_color='lightgrey', node_size=500)
plt.axis('off')
plt.show()
```

# 車輪グラフ

```python
n = 10
G = nx.wheel_graph(n)
p = nx.spring_layout(G, iterations=100)
nx.draw_networkx(G, pos=p,
                 node_color='lightgrey', node_size=500)
plt.axis('off')
plt.show()
```

# 格子グラフ

```python
m,n = 4,3
G=nx.grid_2d_graph(m,n)
# 辞書形式で渡してもよい
p = {v:v for v in G.nodes()}
nx.draw_networkx(G,pos=p,node_color='lightgrey',node_size=500)
plt.axis('off')
plt.show()
```

# 立方体グラフ

```python
n = 3
G = nx.nx.hypercube_graph(n)
p = nx.spring_layout(G, iterations=100)
nx.draw_networkx(G, pos=p,
                 node_color='lightgrey', node_size=500)
plt.axis('off')
plt.show()
```

```python
n = 5
G = nx.nx.hypercube_graph(n)
p = nx.spring_layout(G, iterations=100)
nx.draw_networkx(G, pos=p,
                 node_color='lightgrey', node_size=500, with_labels=False)
plt.axis('off')
plt.show()
```

# 次数

```python
G = nx.random_geometric_graph(100,0.1)

print('次数の合計:', sum(nx.degree(G,v) for v in G.nodes())) # 接続している枝の数の合計
print('枝の数の2倍:', 2*len(G.edges())) # 枝の数の2倍と一致する
print('奇点の数:', len([v for v in G.nodes() if nx.degree(G,v)%2 == 1])) # 枝が奇数しか接続していない頂点は必ず偶数個になる
```

# 同型

```python
G1 = nx.Graph()
G1.add_edges_from([(1,4),(1,5),(1,6),(2,4),
               (2,5),(2,6),(3,4),(3,5),(4,6)])
G2 = nx.Graph()
G2.add_edges_from([('a','b'),('a','d'),('a','f'),('b','c'),
               ('a','e'),('c','d'),('c','f'),('d','e'),('e','f'),])
G3 = nx.Graph()
G3.add_edges_from([('x','z'),('x','y'),('y','z'),('x','u'),
               ('y','v'),('z','w'),('u','v'),('v','w'),('w','u'),])

print(nx.is_isomorphic(G1,G2))
print(nx.is_isomorphic(G1,G3))
```

# 部分グラフ

```python
# 頂点1,3,4が誘導する部分グラフを作成
G = nx.Graph()
G.add_edges_from([(1,2),(1,3),(1,4),(2,4),(3,4)])
G1 = G.subgraph((1,3,4))
nx.draw_networkx(G1, node_color='lightgray')
plt.axis('off')
plt.show()
```

# 連結グラフ

```python
G = nx.path_graph(4)
G.add_path([10, 11, 12])
print(nx.is_connected(G)) # 連結グラフかどうか確認
for c in nx.connected_components(G): # 連結成分に分解
    print(c)
```

```python
nx.draw_networkx(G, node_color='lightgray')
plt.axis('off')
plt.show()
```
