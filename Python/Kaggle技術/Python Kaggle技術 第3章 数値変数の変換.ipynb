###### Log

```python
import numpy as np
x=np.array([1,10,100,1000,10000])

#単に対数を取る
x1=np.log(x)
print(x1)

#1を加えた後に対数をとる(値が0になるときのため)
x2=np.log1p(x)
print(x2)

#絶対値の対数を取ってから元の符号を付加する
x3=np.sign(x)*np.log(np.abs(x))
print(x3)
```

### 3.5.3 feature hashing

```python
df=pd.DataFrame({'Category':['A1','B1','C1','D1','E1'],
                 'Category2':['A2','B2','C2','D2','E2']})
df.head()
```

```python
#Onehotencodingの水準数を減らした変換

from sklearn.feature_extraction import FeatureHasher

#カテゴリ変数をループしてfeature hashing
for c in columns:
    fh=FeatureHasher(n_features=3,input_type='string',alternate_sign=False) #水準数を指定
    #変数を文字列に変換してからFeatureHasherを適用
    hash_data=fh.transform(df[[c]].astype(str).values)
    #データフレームに変換
    #疎行列が返ってきているので、変換が必要
    hash_df=pd.DataFrame(hash_data.todense(),columns=[f'{c}_{i}' for i in range(3)])
    #元のデータフレームと結合
    df=pd.concat([df,hash_df],axis=1)

df.head() 
```

### 3.5.5 target encoding
###### 決定木に有効な手法
###### 目的変数の値に相関するように、カテゴリ変数を並べ変える

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

folder='C:\\Users\\nakam\\Python\\Data\\kaggle'
os.chdir(folder)
df=pd.read_csv('train.csv')
df.head()
```

```python
df=df[['Embarked','Sex','Survived']]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test=train_test_split(
df,random_state=1)
```

```python
from sklearn.model_selection import KFold

#学習データ全体で各カテゴリにおけるtargetの平均を計算
for c in ['Embarked','Sex']:
    target_mean=X_train.groupby(c)['Survived'].mean()

    #テストデータのカテゴリを置換
    X_test[c]=X_test[c].map(target_mean)
    
    #学習データの変換後の値を格納する配列おｗ準備
    tmp=np.repeat(np.nan,X_train.shape[0]) #nanをX_train.shape[0]個繰り返す
    
    #学習データを分割
    #n_splits=4-9がいいと思われる
    kf=KFold(n_splits=4,shuffle=True,random_state=72)
    for idx_1,idx_2 in kf.split(X_train):
        #out of foldで各カテゴリにおける目的変数の平均を計算
        target_mean=X_train.iloc[idx_1].groupby(c)['Survived'].mean()
        #変換後の値を一時配列に格納
        tmp[idx_2]=X_train[c].iloc[idx_2].map(target_mean)
    
    #変換後のデータで元の変数を置換
    X_train[c]=tmp  
```

```python
X_test.head()
```

```python
X_train.head()
```

```python
#クロスバリデーションを実施する場合、vaalidationデータ以外でtarget encodingする必要がある
#クロスバリデーションのfold毎に変換をかけなおす
from sklearn.model_selection import KFold

#クロスバリデーションのfoldごとにtarget encodingをやり直す
kf=KFold(n_splits=4,shuffle=True,random_state=71)
for i,(tr_idx,va_idx) in enumerate(kf.split(X_train)):
    
    #学習データとバリデーションデータを分ける
    tr_x,va_x=X_train.iloc[tr_idx].copy(),X_train.iloc[va_idx].copy()
    
    #変数をループしてtarget encoding
    for c in ['Embarked','Sex']:
        #学習データ全体で各カテゴリにおけるtargetの平均を計算
        target_mean=tr_x.groupby(c)['Survived'].mean()

        #テストデータのカテゴリを置換
        va_x[c]=va_x[c].map(target_mean)
        
        #学習データの変換後の値を格納する配列を準備
        tmp=np.repeat(np.nan,tr_x.shape[0])
        kf_encoding=KFold(n_splits=4,shuffle=True,random_state=72)
        for idx_1,idx_2 in kf_encoding.split(tr_x):
            #out-of-foldで各カテゴリにおける目的変数の平均を計算
            target_mean=tr_x.groupby(c)['Survived'].mean()
            #変換後の値を一時配列に格納
            tmp[idx_2]=tr_x[c].iloc[idx_2].map(target_mean)
            
        tr_x.loc[:,c]=tmp
    print(tr_x,va_x)
```

## 3.11 次元削減、教師無し学習
### 3.11.1 主成分分析

```python
#PCAを適用して、可視化する
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
cancer=load_breast_cancer()
y=cancer.target
```

```python
from sklearn.preprocessing import StandardScaler

scale=StandardScaler()
X_scaled=scale.fit_transform(cancer.data)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
for i in np.unique(y):
    temp=X_pca[y==i]
    plt.scatter(temp[:,0],temp[:,1])
```

### 3.11.2 非負値行列因子分解(NMF)

```python
from sklearn.preprocessing import MinMaxScaler

#非負しか扱えないので変換
minmax=MinMaxScaler()
X_scaled=minmax.fit_transform(cancer.data)
```

```python
from sklearn.decomposition import NMF
nmf=NMF(n_components=2,init='random',random_state=71)
X_nmf=nmf.fit_transform(X_scaled)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
for i in np.unique(y):
    temp=X_nmf[y==i]
    plt.scatter(temp[:,0],temp[:,1])
```

### 3.11.4 線形判別分析(LDA)

```python
for i in np.unique(y):
    temp=X_lda[y==i]
    plt.scatter(temp[:,0],np.zeros(len(temp)))
```
